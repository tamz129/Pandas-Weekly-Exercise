```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
# Membuat kategori baru menurut threshold
d ={'hobi':['jogging','mancing','renang',
            'mancing','mancing','baca',
            'baca','mancing','fotografi',
            'mancing','camping']}

df = pd.DataFrame(d)
df
```

```python
df['hobi'].value_counts()
```

```python
persentase = df['hobi'].value_counts(normalize=True) #untuk menghitung jumlah persentase setiap value
persentase
```

```python
threshold = 0.1
hobi_lain = persentase [persentase < threshold].index #untuk menyatukan sejumlah hobi yang proporsinya dibawah 10% menjadi kategori lain atau hobi lain dan nilainya akan disimpan berupa nilai labelnya
hobi_lain
```

```python
df['hobi'] = df['hobi'].replace(hobi_lain, 'lainnya') #untuk mengganti hobi 'jogging', 'renang', 'fotografi', dan 'camping' menjadi lainnya
df['hobi']
```

```python
df['hobi'].value_counts(normalize=True)
```
