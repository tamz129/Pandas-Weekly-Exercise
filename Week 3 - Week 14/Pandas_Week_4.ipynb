```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
# Konversi nilai numerik ke dalam kategori pada Data Frame
n_rows = 10
n_cols = 1
cols= ('usia',)
df = pd.DataFrame(np.random.randint(1, 99, size=(n_rows, n_cols)),
columns=cols)

df
```

```python
df['kelompok_usia'] = pd.cut (df['usia'],
                             bins = [0, 18, 65, 99],
                             labels = ['anak', 'dewasa', 'manula'])
df
```

```python
# Menggabungkan dua Pandas Data Frame
n_rows = 5
n_cols = 5
cols= tuple('ABCDE')
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)

df.head()
```

```python
df1 = df.copy(deep = True)
df1 = df1.drop([1, 4])
df1
```

```python
df2 = df.copy(deep = True)
df2= df2.drop( [0, 3])
df2
```

```python
df_inner = pd.merge(df1, df2, how='inner')
df_inner
```

```python
df_outer = pd. merge(df1, df2, how='outer')
df_outer
```

```python
#  Memecah nilai string suatu kolom ke dalam beberapa kolom baru pada Pandas Data Frame
data = {'nama':['Didi Kempot', 'Glenn Fredly', 'Mbah Surip'],
        'tempat kelahiran':['Surakarta, Jawa Tengah', 'Jakarta, DKI Jakarta', 'Mojoker, Jawa Timur']}
df = pd.DataFrame(data)
df
```

```python
df[['nama_depan', 'nama_belakang']] = df['nama'].str.split(' ', expand=True)
df
```

```python
df[['kota', 'propinsi']] = df['tempat kelahiran'].str.split(',', expand=True)
df
```
