### 01: Menyertakan Prefix dan Suffix pada seluruh Kolom Data Frame
Import Modules

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
# Menyertakan Prefix dan Suffix pada seluruh Kolom Pandas Data Frame
n_rows = 5
n_cols = 3
cols= tuple('ABCDE')
df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)),
                  columns=cols)

df
```

```python

tuple('ABCDE')
```

```python
df.add_prefix('kolom_')
```

```python

df.add_suffix('_field')
```

```python

# Pemilihan baris (rows selection) pada Data Frame
n_rows = 10
n_cols = 5
cols= tuple('ABCDE')
df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)),
                  columns=cols)

df
```

```python
df[(df['A'] == 1) | (df['A'] == 3)]
```

```python
df[df['A']. isin([1, 3])]
```

```python
df[~df['A' ].isin([1, 3])]
```

```python
# Konversi tipe data String ke Numerik pada kolom Data Frame
data = {'coll': ['1', '2', '3', 'teks'],
        'col2': ['1', '2', '3', '4']}

df = pd.DataFrame (data)
df
```

```python
 df.dtypes
     
```

```python

df_x = df.astype({'col2':'int'})
df_x
```

```python

df_x.dtypes
```

```python

df.apply(pd.to_numeric, errors='coerce')
```

```python

n_rows = 5
n_cols = 2
cols = ['bil pecahan', 'bil_bulat']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)

df['bil pecahan'] = df['bil pecahan'].astype('float')

df.index = pd.date_range(start='2023-01-01', periods=n_rows, freq='H')

df = df.reset_index()
df['teks'] = list('ABCDE')

df
```
